```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
import sklearn.svm
```

```python
path = r"G:\AI学习资料\machine-learning-2014\机器学习课程2014源代码\python代码\ex6-SVM\data\ex6data1.mat"
raw_data = loadmat(path)
```

```python
def get_data(raw_data):
    X,y = raw_data["X"], raw_data["y"]
    return X, y.flatten()
X, y = get_data(raw_data)
print(f'X shape:{X.shape}, y shape:{y.shape}')
```

```python
def get_theta(num):
    return np.zeros(num)
```

### Plot raw data

```python
fig,axes = plt.subplots(figsize=(8,6))
axes.scatter(X[:,0], X[:,1], c=y.flatten(), s=50)
axes.set_title("raw_data")
axes.set_xlabel("x1")
axes.set_ylabel("x2")
```

### Try C = 1

```python
sv1 = sklearn.svm.LinearSVC(C=1, loss='hinge', max_iter=20000)
sv1.fit(X, y)
sv1.score(X,y)
```

```python
sv1_dec = sv1.decision_function(X)
fig, ax = plt.subplots(figsize=(8,6))
ax.scatter(X[:,0], X[:,1], s=50, c=sv1_dec, cmap='RdBu')
ax.set_title('SVM (C=1) Decision Confidence')
```

```python
svm100 = sklearn.svm.LinearSVC(C=100, loss='hinge',max_iter=20000)
svm100.fit(X,y)
svm100.score(X,y)
```

```python
svm100_dec = svm100.decision_function(X)
fig, ax = plt.subplots(figsize=(8,6))
ax.scatter(X[:,0], X[:,1], s=50, c=svm100_dec, cmap='RdBu')
ax.set_title('SVM (C=100) Decision Confidence')
```
