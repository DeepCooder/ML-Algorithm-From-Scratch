
**Exercise 1 - Support Vector Machines**

This exercise covers training support vector machines (SVMs) on two-dimensional (2D) datasets to classify data into positive and negative examples. It involves:

- Implementing a Gaussian kernel function to allow the SVM to learn non-linear decision boundaries
- Training an SVM with different C parameters on the example datasets to see the effects on the decision boundary and outliers
- Using a cross validation set to select the best C and σ parameters for an SVM with a Gaussian kernel on one of the example datasets

**Theory and Functions**

Relevant theory includes the SVM objective and use of kernels to map inputs to a higher dimensional feature space to learn non-linear decision boundaries.

Key functions to implement:

- `gaussianKernel(x, y)`: Computes Gaussian kernel similarity between two examples 
- Find best `C` and `σ` parameters by training SVM models on a cross validation set

The submitted solutions include:

- `gaussianKernel `: Gaussian kernel implementation
- `dataset3Params `: Best parameters C and σ for Dataset 3

The datasets provided are:

- `ex6data1 at`, `ex6data2 at`, `ex6data3 at`: Example datasets
- `Xval`, `yval`: Cross validation data for Dataset 3

**Steps**

1. Implement Gaussian kernel 
2. Train SVM using `svmTrain` for different `C` values
3. Use cross validation to find best parameters C and σ
4. Update `dataset3Params ` with best parameters

**Exercise 2 - Spam Classification**

This exercise covers training an SVM classifier to categorize emails as spam or not spam. It involves:

- Preprocessing emails by lowercasing, stripping HTML, normalizing URLs/numbers etc.
- Extracting feature vectors from emails based on word occurrences 
- Training an SVM classifier on the email features
- Evaluating predictor words that are most indicative of spam

**Theory and Functions**  

Relevant theory includes text processing pipelines and feature extraction for SVM classification.

Key functions to implement:

- `processEmail(str)`: Preprocess email text 
- `emailFeatures(word_indices)`: Create feature vector from word indices

The submitted solutions include:

- `processEmail `: Email preprocessor
- `emailFeatures `: Email feature extractor

The datasets:

- `spamTrain at`, `spamTest at`: Spam classifier training and test data
- `vocab.txt`: Vocabulary list

**Steps**

1. Preprocess emails with `processEmail`
2. Extract email feature vectors with `emailFeatures` 
3. Train SVM classifier on email features 
4. Analyze most predictive words for spam
