```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
import sklearn.svm
```

```python
# Define a function 'gaussion' to compute the Gaussian kernel between two vectors 'xi' and 'xj'.
# 'xi' and 'xj' are the input vectors.
# 'sigma' is the bandwidth of the Gaussian kernel.
def gaussion(xi,xj,sigma):
    # Compute the difference 'x' between 'xi' and 'xj'.
    x = xi - xj
    # Compute the inner product 'inner' of 'x' with itself.
    inner = x@x
    # Compute the Gaussian kernel 'res' using the formula exp(-(inner/(2*sigma**2))).
    res = np.exp(-(inner/(2*sigma**2)))
    # Return 'res'.
    return res
```

```python
x1 = np.array([1, 2, 1])
x2 = np.array([0, 4, -1])
sigma = 2
gaussion(x1,x2,sigma)
```

```python
path = r"D:\MyProjects\Python\Andrew Ng_ML_Exercise\python代码\ex6-SVM\data\ex6data2.mat"
data = loadmat(path)
```

```python
def getxy(data):
    X = data["X"]
    y = data["y"].flatten()
    return X,y
X,y = getxy(data)
```

### plot raw data

```python
fig,ax = plt.subplots()
ax.scatter(x=X[:,0],y=X[:,1],c=y)
ax.set_title("raw data")
ax.set_xlabel("x1")
ax.set_ylabel("x2")
plt.plot()
```

### impliment gaussion model

```python
svc = sklearn.svm.SVC(C=100,kernel='rbf', gamma=10, probability=True)
svc.fit(X,y)
svc.score(X,y)
```

```python
predict_proba = svc.predict_proba(X)[:,0]
```

```python
fig, ax = plt.subplots()
ax.scatter(X[:,0],X[:,1],s=30,c=predict_proba,cmap='Reds')
```
