```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
path = r"D:\MyProjects\Python\Andrew Ng_ML_Exercise\python代码\ex7-kmeans and PCA\data\bird_small.png"
raw_data = plt.imread(path)
width = raw_data.shape[0]
high = raw_data.shape[1]
```

```python
data = raw_data.reshape(width*high, 3)
data.shape
```

```python
k=16
km = KMeans(n_clusters=k)
km.fit(data)
c_i = km.predict(data)
cet = km.cluster_centers_
```

```python
cop_data = cet[c_i]
```

```python
cop_data = cop_data.reshape(width,high,3)
```

```python
fig,axes = plt.subplots(1,2,figsize=(6,4))
axes[0].imshow(raw_data)
axes[1].imshow(cop_data)
axes[0].set_title('Orignal')
axes[1].set_title(f'Compressed with {k} colors')
plt.show()
```
